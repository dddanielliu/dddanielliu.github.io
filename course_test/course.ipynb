```python
def update_column(row):
    if row[df.keys()[0]].startswith("002"):
        return "體育通識"
    if row[df.keys()[0]].startswith("003"):
        return "國防軍訓"
    if row[df.keys()[0]].startswith("031"):
        return "中文通識"
    if row[df.keys()[0]].startswith("032"):
        return "外文通識"
    if row[df.keys()[0]].startswith("041"):
        return "人文通識"
    elif row[df.keys()[0]].startswith("042"):
        return "社會通識"
    elif row[df.keys()[0]].startswith("043"):
        return "自然通識"
    elif row[df.keys()[0]].startswith("044"):
        return "跨領域通識(人文、社會)"
    elif row[df.keys()[0]].startswith("045"):
        return "書院通識"
    elif row[df.keys()[0]].startswith("046"):
        return "資訊通識"
    elif row[df.keys()[0]].startswith("090"):
        return "通識"
    else:
        return np.nan

# Apply the function to each row
df['科目類別'] = df.apply(update_column, axis=1)
```

```python
def time_period_filter(s):
    if (s[0] == '一'):
        return (s[1:] in "1234C" or s[1:] in "78EFG")
    elif (s[0] == '二'):
        return (s[1:] in "78EFG")
    elif (s[0] == '三'):
        return False
    elif (s[0] == '四'):
        return (s[1:] in "78EFG")
    elif (s[0] == '五'):
        return (s[1:] in "1234C" or s[1:] in "78EFG")
    else:
        return True
```

```python
def replace_first_line(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()

    # Replace ',' with '","' in the first line
    lines[0] = lines[0].replace(',', '","')

    with open(file_path, 'w') as file:
        file.writelines(lines)

# Replace the first line in the file
file_path = 'output.csv'  # Replace with your actual file path
replace_first_line(file_path)
```
